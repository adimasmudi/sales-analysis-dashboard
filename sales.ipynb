```python
import pandas as pd

df = pd.read_csv('sales.csv')

df.head()
```

```python
df_new = df.iloc[:,:55]
df_new.head()
```

```python
df_new.shape
```

```python
x_label = df_new.columns[1:-2]
x_label
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = df_new.loc[df_new['Product_Code'] == 'P1']
print(data)
```

```python
def get_data_by_productCode(code):
    data = {'week':[],'sales':[]}
    get_data = df_new.loc[df_new['Product_Code'] == code]
    for i in range(len(x_label)):
        
        data['week'].append(i+1)
        data['sales'].append(*get_data[x_label[i]])
    df_prod = pd.DataFrame(data)
    return df_prod

```

```python
def display_by_productCode(code):
    data = get_data_by_productCode(code)
    sns.lineplot(data=data, x='week', y='sales')
display_by_productCode('P6')
```
